**Регрессионный анализ** - можно ли провести линейную зависимость между двумя выборками. Возьмем Ladder - удовлетворение жизнью, и generosity - доброта. Мы хотим найти прямую, которая разделит точки в пространстве наполовину. Это можно сделать графически. Нарисовать график. Можно ли сказать, что данная прямая является собственным вектором? При проецировании данных на эту прямую разброс данных максимален?

```python
ladder = data_clear['Ladder']
generosity = data_clear['Generosity']
sns.regplot(x= ladder, y=generosity, data=data_clear, fit_reg=True)
plt.xlabel('удовлетворение жизнью')
plt.ylabel('доброта')
plt.title('Корреляционная зависимость двух выборок')
```

Исходя из свойств:
линейная регрессия - прямая минимизирующая расстояние от дочек до неё
собственный вектор - при проецировании всех точек на собственный вектор разброс данных получается максимальным

можно сказать, что это всё-таки 2 разные прямые, при проецировании данных на прямую, деляющую точки в пространстве наполовину, разброс данных не будет максимален.

Ниже иллюстрация, где рисую линию регресии и собственные вектора

```python
ladder = data_clear['Ladder']
generosity = data_clear['Generosity']
sns.regplot(x= ladder, y=generosity, data=data_clear, fit_reg=True)
plt.xlabel('удовлетворение жизнью')
plt.ylabel('доброта')
plt.title('Корреляционная зависимость двух выборок')

import numpy.linalg as la
import math
from sklearn.preprocessing import StandardScaler
mean1 = statistics.mean(ladder)
ladder -= mean1
mean2 = statistics.mean(generosity)
generosity -= mean2
Y = np.vstack((ladder, generosity)).T
w,v = la.eig(np.cov(Y.T))

for i in range(len(w)):
    plt.suptitle('Собственный вектор № '+ str(i+1), fontsize=16)    
    plt.plot(ladder, generosity, 'r.', zorder=1)
    plt.plot([mean1, mean1+w[i]*v[:,i][0]], [mean2, mean2 +w[i] * v[:,i][1]],color='red', linewidth=3)

axes = plt.gca()
axes.set_xlim([-80,80])
axes.set_ylim([-80,80])
plt.grid(False)
plt.show()
```
