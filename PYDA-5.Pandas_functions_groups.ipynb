```python
import pandas as pd
```

```python
#Task 1
ratings = pd.read_csv("ratings.csv")
```

```python
def ratings_classify(rating):
    if rating <= 2:
        return 'low class'
    elif 2 < rating <= 4:
        return 'mid class'
    elif rating == 4.5 or rating == 5:
        return 'high class'
    else:
        return 'unknown'
```

```python
ratings["class"] = ratings["rating"].apply(ratings_classify)
ratings.head(10)
```

```python
#task 2
keywords = pd.read_csv("keywords.csv")
geo_data = {

'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo_classify(keyword):
    region = 'undefined'
    for key, value in geo_data.items():
        if set(keyword.split(' ')) & set(value):
            region = key
    return region

keywords["region"] = keywords["keyword"].apply(geo_classify)

#test
print(keywords[keywords.keyword.str.contains('псков')].head())
print(keywords[keywords.keyword.str.contains('тула')].head())
```

```python
#task 3
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
years = list(range(1950,2010))
```

```python
def production_year(title):
    year = '1900'
    for year in years:
        if title.find(str(year)) != -1:
            return str(year)
```

```python
movies["year"] = movies["title"].apply(production_year)
movies.head(20)
joined = movies.merge(ratings, how ='inner', on ='movieId')
joined.head()
```

```python
joined[['year', 'rating']].groupby('year').mean().sort_values('rating', ascending = False)
```
